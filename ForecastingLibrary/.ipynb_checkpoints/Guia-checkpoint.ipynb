### Uso de las funciones

Todas las funciones requieren una matriz de 10 columnas y al menos 4 filas como parámetro. Las columnas de esta matriz hacen referencia respectivamente a: Hora de la medición (formato de 24 horas), Temperatura Ambiente (°C), Temperatura del Estanque (°C), Oxígeno Disuelto (mg/L), pH, Conductividad Eléctrica (uS/cm), Amonio sin ionizar NH3 (mg/L), Amonio ionizado NH4 (mg/L), Nitritos (mg/L) y Alcalinidad (mg/L). Cada fila corresponde a una ronda completa de mediciones que debe estar organizada cronológicamente de tal manera que la última fila de la columna contenga la última medició realizada por cada parámetro y la hora de la misma. No pueden haber espacios sin información, de ser así, aproximar este valor antes de utilizar las funciones. 


A continuación se presenta un código para adecuar los datos ya tomados a las funciones, sin embargo este código no es necesario para utilizar los datos que ingresan los piscicultores. Solamente se utiliza por la forma específica en que yo guardé mis datos.

```python
# Carga de datos para ser usado en la predicción
mat = scipy.io.loadmat('Datos2.mat')
M=mat.get("M")
Z=[]
for i in range(len(M)):
    if M[i,-1]!=0:
        Z.append(M[i,:])
Z=np.array(Z)
Z=np.delete(Z,[1,2], axis=1)
X=np.delete(Z,-1,axis=0)
X=np.delete(X,range(1,9),axis=1)
print(X)
```

#### Predicción Temperaturas

```python
TempAmb=predTempAmb(X)
print('Temperatura Ambiente: '+ str(np.round(TempAmb,decimals=2)) +'°C')
TempEst=predTempEst(X)
print('Temperatura del Estanque: '+ str(np.round(TempEst,decimals=2)) +'°C')
```

#### Predicción Oxígeno Disuelto

```python
OD=predOD(X)
print('Oxígeno Disuelto: '+ str(np.round(OD,decimals=3)) +' mg/L')
```

#### Predicción pH

```python
PH=predPH(X)
print('pH: '+ str(np.round(PH,decimals=3)))
```

#### Predicción Conductividad

```python
CondElec=predCondElec(X)
print('Conductividad eléctrica: '+ str(np.round(CondElec,decimals=1)) +' uS/cm')
```

#### Predicción Ciclo del Nitrógeno

```python
NH3=predNH3(X)
print('Amonio sin ionizar NH3: '+ str(np.round(NH3,decimals=3)) +' mg/L')
NH4=predNH4(X)
print('Amonio ionizado NH4: '+ str(np.round(NH4,decimals=3)) +' mg/L')
Nit=predNit(X)
print('Nitritos: '+ str(np.round(Nit,decimals=3)) +' mg/L')
```

#### Predicción Alcalinidad

```python
Alk=predAlk(X)
print('Alcalinidad: '+ str(np.round(Alk,decimals=2)) +' mg/L')
```

### Predicción Global de todas las variables con una ventana V hacia el futuro

Esta última función predice todas las variables al mismo tiempo posibilitando incrementar la ventana de predicción todo lo que se desee. Además de la matriz anteriormente mencionada, esta función también debe recibir como parámetro la cantidad **v** de mediciones que se desean predecir hacia el futuro. Por ejemplo si se quieren las mediciones del día siguiente v debe ser igual a 2, o si se desean las de la semana siguiente v debe ser 14.

```python
X_pred=predTotal(X,14)
print(X_pred)
```

Esta matriz contiene toda la información necesaria, por ejemplo, si se desea conocer la temperatura del estanque de la próxima semana, se accede a ella seleccionando la tercera columna de esta matriz. Recordar que el índice en python inicia en 0.

```python
TemEst=X_pred[:,2]
print('Temperatura futura del estanque: '+ str(np.round(TemEst,decimals=2)))
```

Si se desea conocer el pH futuro se accede a dicha columna (La columna 5 en este caso)

```python
PH=X_pred[:,4]
print('pH futuro: '+ str(np.round(PH,decimals=2)))
```

Además, nunca separar modelos y librerías, siempre deben estar en el mismo directorio para que la librería pueda utilizarlos.
